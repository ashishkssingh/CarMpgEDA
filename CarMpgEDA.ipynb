## Plots describing the data

### Plotting all the values in dataframe

```python
sns.pairplot(df)
```

## Plotting features showing correlation to mpg

### Acceleration to mpg with cylinder as category

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='mpg',y='acceleration',hue='cylinders',data=df)
```

- **Cylinders have a direct relation with mpg, higher the cylinder gives lower mpg.**
- **Cylinder to mpg moderately related and directly proportional**

### Weight to mpg with cylinder as category

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='mpg',y='weight',hue='cylinders',data=df)
```

- **Heavier cars requires more cylinders and hence resulting in lower mpg.**
- **Weight to mpg highly related and inversely proportional**

### Displacement to mpg with cylinder as category

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='mpg',y='displacement',hue='cylinders',data=df)
```

- **lower cylinders have lower displacement and higher mpgs**
- **displacement to mpg inversely correlated**

### Model year to mpg with cylinder as category

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='mpg',y='model year',hue='cylinders',data=df)
```

- **mpg has improved over the years**  
- **Moderate relation between year and mpg though**

### Cylinder to mpg to see relationship and distribution

```python
plt.figure(figsize=(10,8))
sns.violinplot(x='cylinders',y='mpg',data=df)
```

- **Higher cylinders give lower mpg**
- **5 cylinder cars give most variance in mpg**

## Confirming insights using correlation matrix

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True)
```

**Insights supported by correlation matrix**:
1. Cylinders, Displacement and weight showing high correlation to mpg.
2. Acceleration and Model year are moderately correlated.

**Caution:** Always remember "correlation does not imply causation".
